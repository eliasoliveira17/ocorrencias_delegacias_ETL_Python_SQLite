```python
# criar a tabela tabLopBPM no BDODS
query_tbLogAreaBPM = '''

    CREATE TABLE IF NOT EXISTS tbLogAreaBPM
    (    
    codBPM INTEGER,
    areaBPM REAL (5, 2),
    dtCarga DATETIME 
    )
'''

# executa a query de criação da tabela
SQL_ODS.execute(query_tbLogAreaBPM)

query_idx_tbLogAreaBPM_codBPM = "CREATE INDEX IF NOT EXISTS idx_tbLogAreaBPM_codBPM ON tbLogAreaBPM (codBPM)"

# executa a query de criação do index
SQL_ODS.execute(query_idx_tbLogAreaBPM_codBPM)

# não há necessidade de confirmar a transação
```

# COLETANDO DADOS PARA DIMENSÃO DE BPM (dBPM)

```python
# inserindo a query

qry_sql = '''
SELECT
    codBPM,
    nmBPM,
    endereco,
    areaBPM
FROM (
SELECT
    a.codBPM,
    a.nmBPM,
    a.endereco,
    b.areaBPM,
    max(a.dtCarga)
FROM tbLogBPM a
JOIN tbLogAreaBPM b
on a.codBPM = b.codBPM
GROUP BY
    a.codBPM,
    a.nmBPM,
    a.endereco,
    b.areaBPM
    )
'''

# leitura de dBPM
dBPM = pd.read_sql(qry_sql, conexaoBDODS)

# confirma a transação
conexaoBDODS.commit()

# encerra conexão com o banco de dados BDODS
# conexaoBDODS.close()

dBPM
```

# CONECTANDO AO DW 

```python
# estabele conexão com BDDW
conexaoBDDW = sql.connect(BDDW)
# definição da manipulação de dados em SQL
SQL_DW = conexaoBDDW.cursor()
```

# CRIANDO dBPM em DW

```python
# query de criação de dBPM
query_dBPM = '''
    CREATE TABLE IF NOT EXISTS dBPM
    (
    idBPM INTEGER PRIMARY KEY AUTOINCREMENT,
    codBPM INTEGER,
    nmBPM VARCHAR (7),
    endereco VARCHAR (200),
    areaBPM REAL (5,2)    
    )
'''

# executar query de criação de dBPM
SQL_DW.execute(query_dBPM)

# query de definição do índice idBPM
query_idx_dBPM_idBPM = "CREATE INDEX IF NOT EXISTS idx_dBPM_idBPM ON dBPM (idBPM)"

# executar query de criação de idBPM
SQL_DW.execute(query_idx_dBPM_idBPM)

# query de definição do índice codBPM
query_idx_dBPM_codBPM = "CREATE INDEX IF NOT EXISTS idx_dBPM_codBPM ON dBPM (codBPM)"

# executar query de criação de idBPM
SQL_DW.execute(query_idx_dBPM_codBPM)
```

# DELETANDO DADOS ANTIGOS E REINICIANDO O AUTOINCRIMENTAL DA CHAVE

```python
# 
SQL_DW.execute('DELETE FROM dBPM')
               
# confirmar transação
conexaoBDDW.commit()
               
# reiniciar o autoincremento do índice
SQL_DW.execute("UPDATE sqlite_sequence SET seq=0 where name='dBPM'")

# confirmar transação
conexaoBDDW.commit()
```

# INSERINDO DADOS NA DIMENSÃO BPM NO DW

```python
# inserindo registros na tabela dDP
print('Iniciando a inserção de dados na tabela dBPM')

# OBS: não há necessidade de informar a coluna de índice (autoincrimental)
SQL_DW.executemany('''INSERT INTO dBPM (codBPM,nmBPM,endereco,areaBPM) VALUES (?,?,?,?)''',dBPM.values.tolist())

# confirma a transação
conexaoBDDW.commit()

print('Carga de', len(dBPM), 'registros em dBPM finalizada ')

```
